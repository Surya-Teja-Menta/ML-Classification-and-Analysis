```python
n=df.columns[5:40]
scaled=df.copy()
```

```python
features=scaled[n]
```

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler().fit(features.values)
```

```python
features=s.transform(features.values)
```

```python
scaled[n]=features
scaled.head()
```

```python
def gend(x):
    if x=='M':
        return 1
    if x=='F':
        return 2
    if x=='not disclosed':
        return 3
```

```python
scaled['gender']=scaled['gender'].apply(gend)
scaled.gender.head()
```

```python
from sklearn.cluster import KMeans
km=KMeans(n_clusters=10,random_state=0,n_jobs=-1)
model=km.fit(scaled)
```

```python
wcss=[]
for i in range(1,20):
    kms=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kms.fit(scaled)
    wcss.append(kms.inertia_)
plt.plot(range(1,20),wcss)
plt.title('The Elbow Curve')
plt.xlabel('The Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
kms=KMeans(n_clusters=10)
kms.fit(scaled)
```

```python
len(kms.labels_)
```

```python
df['cluster']=kms.labels_
df.head()
```

```python
plt.figure(figsize=(12,7))
axis = sns.barplot(x=np.arange(0,10,1),y=df.groupby(['cluster']).count()['age'].values)
x=axis.set_xlabel("Cluster Number")
x=axis.set_ylabel("Number of Students")
```
