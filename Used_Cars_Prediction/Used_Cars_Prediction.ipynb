```python
exce_odo_mean=df[df['condition']=='excellent']['odometer'].mean()
good_odo_mean=df[df['condition']=='good']['odometer'].mean()
like_odo_mean=df[df['condition']=='like new']['odometer'].mean()
fair_odo_mean=df[df['condition']=='fair']['odometer'].mean()
new_odo_mean=df[df['condition']=='new']['odometer'].mean()
salvage_odo_mean=df[df['condition']=='salvage']['odometer'].mean()

```

```python
print('Like new average odometer:', round( like_odo_mean,2))
print('Excellent average odometer:', round( exce_odo_mean,2))
print('Good average odometer:', round( good_odo_mean,2))
print('Fair average odometer:', round( fair_odo_mean,2))
print('Salvage average odometer:', round( salvage_odo_mean,2))
```

```python
df.loc[df.year>=2020,'condition']=df.loc[df.year>=2020,'condition'].fillna('new')
df.loc[df.year>=2021,'condition']=df.loc[df.year>=2021,'condition'].fillna('new')
```

```python
df.loc[df['odometer'] <= like_odo_mean, 'condition'] = df.loc[df['odometer'] <= like_odo_mean, 'condition'].fillna('like new')

df.loc[df['odometer'] >= fair_odo_mean, 'condition'] = df.loc[df['odometer'] >= fair_odo_mean, 'condition'].fillna('fair')

df.loc[((df['odometer'] > good_odo_mean) & 
       (df['odometer'] <= exce_odo_mean)), 'condition'] = df.loc[((df['odometer'] > good_odo_mean) & 
       (df['odometer'] <= exce_odo_mean)), 'condition'].fillna('excellent')

df.loc[((df['odometer'] > like_odo_mean) & 
       (df['odometer'] <= good_odo_mean)), 'condition'] = df.loc[((df['odometer'] > like_odo_mean) & 
       (df['odometer'] <= good_odo_mean)), 'condition'].fillna('good')

df.loc[((df['odometer'] > good_odo_mean) & 
       (df['odometer'] <= fair_odo_mean)), 'condition'] = df.loc[((df['odometer'] > good_odo_mean) & 
       (df['odometer'] <= fair_odo_mean)), 'condition'].fillna('salvage')
```

```python
from sklearn import datasets,linear_model
from sklearn.model_selection import train_test_split
from matplotlib import pyplot as plt
```

```python
y=df.price
x=df.drop('price',1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.ensemble import RandomForestRegressor as rfr
reg=rfr(n_estimators=20,random_state=0)
reg.fit(x_train,y_train)
y_pred=reg.predict(x_test)
```

```python
from sklearn import metrics
from sklearn.metrics import *
print('Mean Absolute Error:', round(metrics.mean_absolute_error(y_test, y_pred),2))
print('Mean Squared Error:', round(metrics.mean_squared_error(y_test, y_pred),2))
print('Root Mean Squared Error:', round(np.sqrt(metrics.mean_squared_error(y_test, y_pred)),2))
print(reg.score(x_test,y_test)*100)
```
